```python
import sys
sys.path.insert(0, "src")

from astrosim.scenario import load_and_build
from astrosim.export.formats import export_json
```

```python
sim = load_and_build("scenarios/lunar_base.yaml")
result = sim.run()
print(f"Energy net: {result.energy_budget.net_kwh:.1f} kWh")
print(f"Steps: {len(result.history)}")
```

```python
export_json(result, "output/tutorial_results.json")
print("Exported to output/tutorial_results.json")
```
